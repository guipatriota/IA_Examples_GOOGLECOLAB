# Gráfico de grupos:

```
import pandas as pd
from sklearn.datasets import make_blobs
dados, _ = make_blobs(n_samples = 1000, n_features = 2, random_state = 7)
```

```
dados = pd.DataFrame(dados, columns = ['coluna1', 'coluna2'])
dados.head()
```

```
import matplotlib.pyplot as plt
plt.scatter(x=dados.coluna1, y=dados.coluna2, s=2)
```

```
dados = pd.DataFrame(dados, columns = ['coluna1', 'coluna2'])
dados.head()
```

```
from sklearn.cluster import KMeans
modelo = KMeans(n_clusters=3)
grupos = modelo.fit_predict(dados)
```

```
plt.scatter(x=dados.coluna1, y=dados.coluna2, 
            c=grupos,
           cmap='viridis', s=2)
```

```
centroides = modelo.cluster_centers_
plt.scatter(dados.coluna1, dados.coluna2,
            c=grupos,
           cmap='viridis',
           s = 2)
plt.scatter(centroides[:, 0], centroides[:, 1],
           marker='X', s=50, linewidths=2,
           color='g', zorder=8)
```

#Exemplo de agrupamento sem sabermos o número exato de clusters necessários - DBSCAN
Ver site: https://www.naftaliharris.com/blog/visualizing-dbscan-clustering/

```
from sklearn.cluster import DBSCAN
```

```
modelo = DBSCAN()
grupos = modelo.fit_predict(dados)
```

```
plt.scatter(dados.coluna1, dados.coluna2,
            c=grupos,
           cmap='viridis')
```
